```python
# Main libraries
import pandas as pd
import numpy as np
import re

# Noebook setup
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 100)
```

```python
# Retrieving data from original CSV file
people = pd.read_csv('../raw_data/people_analytics.csv')

people.head()
```

```python
def camel_case_split(str): 
    ''' 
    Splitting CamelCase with underscore separators.
    
    '''
    return '_'.join(re.findall(r'[A-Z](?:[a-z]+|[A-Z]*(?=[A-Z]|$))', str))
```

```python
# Creating a new_columns list to store new columns names
new_columns = []

# Applying camel_case_split function to column name
for str in people.columns.tolist():
    new_columns.append(camel_case_split(str).lower())
    
# Replacing column names with new_columns info
people.columns = new_columns
```

```python
# Dropping uninformative column 'employee_count'
people.drop(columns = 'employee_count', inplace = True)
```

```python
# Renaming 'employee_number' by 'employee_id' for enhancing comprenhention
people.rename(columns = {'employee_number': 'employee_id'}, inplace = True)
```

```python
# Setting 'employee_id' as dataset index
people.set_index('employee_id', verify_integrity = True, inplace = True)

# View changes
people.head()
```
